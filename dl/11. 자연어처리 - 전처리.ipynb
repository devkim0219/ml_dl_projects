# 사전(단어장, vocab, ...) 생성

## 개요

- 페이스북의 MUSE에서는 사전 구축 방법을 코드와 함께 제공
- 언어별로 단어가 많이 사용되거나 하면 사전이 필수적으로 필요하게 된다.
- 해당 코퍼스를 기준으로 형태소를 뽑아내면(분절) 이를 사전(단어장)으로 구성할 수 있다.
    - 특정 형태소에 대해 인덱스가 부여가 되고
    - 그 인덱스를 기준으로 벡터화를 수행할 수 있다.
    - 단어 임베딩 처리를 통해서 인공신경망에서 훈련할 수 있는 데이터화 처리가 완성된다.

## 단어장 생성 방식

- 중복되지 않는 토큰(고유하다, 사전내에서는)
- 해당 토큰에 고유한 숫자를 부여(인덱스)
- 샘플 데이터는 네이버 감성 영화 코퍼스 데이터 사용
    - https://github.com/e9t/nsmc -> ratings.txt 업로드
    - 컬럼 3개 : id, document, label
    - label은 1이면 긍정, 0이면 부정


```
# step 1 : 코퍼스 수집(git을 통해서 네이버 감성 영화 코퍼스 자료 획득)
# step 2 : 분절용 한글 처리 형태소 분석기(분절 세부 과정은 통으로 분절이라고 표현)
from konlpy.tag import Mecab

tokenizer = Mecab()
```

```
# step 3 : 사전, 단어장 만들기
# step 3-1 : 파일을 읽는다.
with open('/content/ratings.txt', 'r') as f:

    # step 3-2 : 내용은 이미 한 줄로 되어 있어서 문장 단위 분절은 필요 없고 바로 라인별로 분해 -> id는 제거
    # 컬럼 파트는 빼고 사용
    raw = f.read().splitlines()[1:]
    # print(len(raw), raw[:3])

    # 데이터 한 개는 \t 으로 구분되어 있다.
    # -> [['내용1..', 1], ['내용2..', 0], ...]
    data = [line.split('\t')[1:] for line in raw]
    # data = list(map(lambda x: x.split('\t')[1:], raw))
    # print(data[:2])

    # step 3-3 : 토큰화처리, 분절처리 -> morphs()
    # [(['', '', ''], 1), (), (), ...]
    data = list(map(lambda x: (tokenizer.morphs(x[0]), x[1]), data))
    print(data[:1])
```

```
# 데이터가 너무 많아서 샘플링을 해서 진행
# 긍정 : 1, 부정 : 0
# 긍정 리뷰 10개, 부정 리뷰 10개를 샘플링
sampling_data = data[5:5+10] + data[-10:]
print(sampling_data[0], sampling_data[-1])
```

```
# 리스트 구성원의 순서(같은 순서)끼리 묶어서 데이터를 구성
# [([], [], [], ...), (1, 1, 0, 0, ...)]
# zip(*시퀀스데이터)
tmp = list(zip(*sampling_data))
print(tmp)
```

```
# 테스트 토큰화된 데이터만
print(tmp[0][0])
```

```
# 사전화 작업은 토큰을 -> 고유번호 대입
# 전체 데이터를 돌면서 -> 하나의 리뷰를 획득 -> 하나의 토큰을 획득 -> 사전에 있었는지 체크
# -> 없었다면(신규단어) -> 새로운 인덱스를 부여
# -> 새로운 인덱스는 어떻게 계산? 딕셔너리(사전의 자료구조)의 길이값으로 세팅하겠다.

# 전체 데이터
len(tmp[0])
```

```
# 사전
vocab = dict()  # {'토큰': 인덱스}

# 추가부분
# 단어장에 없는 형태소가 있는 경우 : unknown
# 문장의 크기를 맞출 때 필요할 경우 문장의 길이를 동일하게 처리하기 위해 padding 추가
# 토치텍스트에서 시퀀스 구성 후 문장이 임베딩될 때 삽입되는 기호와 동일하게 구성
# <unk>:0, <pad>:1
vocab['<unk>'] = 0
vocab['<pad>'] = 1

for sentence in tmp[0]:
    # print(sentence)

    for token in sentence:
        # print(token)

        if not token in vocab:
            vocab[token] = len(vocab)
            # vocab.setdefault(token, len(vocab))

len(vocab), vocab['-']
```

## 수치화

```
def numericalize(sent, vocab):
    
    # sent에서 token을 하나씩 꺼내서 사전에 대조하여 인덱스(토큰의 인덱스값, 특성값)를 구한다.
    # '사랑' -> 2번으로 표기
    tmp = list()

    for token in sent:

        # 해당 토큰이 사전에 존재하는가?
        if token in vocab:
            tmp.append(vocab[token])
            
        else:
            tmp.append(vocab.get('<unk>'))
            
    return tmp

numericalize_data = [(numericalize(sent, vocab), label) for sent, label in sampling_data]
print(numericalize_data[-2])
print(sampling_data[-2])
```

## 단어 주머니(Bag of Words)

- 고전 기법
- 단어의 출현빈도로만 문장별로 기록(순서는 미고려)
- 문서-단어 행렬이라고 표현 : term-matrix
- 행렬 구성
    - 행 : 문장의 개수
    - 열 : 고유 단어의 개수

```
import torch
from collections import Counter
```

```
# 사용 데이터는 numericalize_data를 사용한다. -> 긍정 10, 부정 10개의 문장
# 문서(장)의 개수 : 행
num_docs = len(numericalize_data)
num_docs
```

```
# 사전의 단어 수 : 열
num_vocabs = len(vocab)
num_vocabs 
```

```
# 행렬 구성 -> 초기값 : 0
term_matrix = torch.zeros(num_docs, num_vocabs)
term_matrix.size()
```

## 단어주머니 문제점

- 문장이 많으면 -> 토큰 수가 늘어나고 -> 수치 벡터의 차원이 커진다.
    - 힙스의 법칙(heaps laws)
    - 로그 함수적으로 증가한다.
- 차원이 크면 특성을 표현하기가 어렵다. -> 차원의 저주 -> 데이터가 희소해진다. -> 0이 많아지고 -> 선형 결합(행렬의 곱)에서 활성화 함수를 통과한 값도 0이 되고 이를 미분(역전파)값도 0이 되고 -> 학습 성과가 저하 -> 단어 벡터의 밀집화(dense)가 필요

## padding

- 문장의 길이는 각각 다르기 때문에 항상 같은 길이로 만들어줘야 효율이 좋다.
- 문장의 최대 길이 기준으로 패딩 처리

```
# 샘플링 데이터 중에 가장 긴 문장의 길이 획득
max_len = max([len(sent) for (sent, _) in numericalize_data])
max_len
```

```
for s, _ in numericalize_data:
    
    # 현재 문장의 길이가 최대 문장의 길이보다 작으면
    c_len = len(s)
    
    if c_len < max_len:

        # padding 표현을 추가한다. -> <pad>
        s += [vocab['<pad>']] * (max_len - c_len)

print(numericalize_data[0])
```

```
# 자연어 처리를 하는데(전처리 -> 레이어 설계 -> 학습, 평가, 결과물(번역, 요약, 분류, ...))
# 엔진 : 파이토치
# 핵심모듈 : 토치텍스트
# 형태소 분석기 : Mecab
# 전처리 과정은 토치텍스트를 사용
from torchtext.data import Field

# 분절 처리를 하는 한글 형태소 분석기는 Mecab
from konlpy.tag import Mecab
```

```
tokenizer = Mecab()
```

```
# ratings.txt의 데이터를 들여다 보면
# id, 리뷰내용, 긍정/부정

# 리뷰 내용에 대한 필드
'''
sequential : 이 필드가 다루는 데이터가 시퀀스 데이터인가? -> 문장(문자열) -> 시퀀스 데이터
use_vocab : 사전을 사용할 것인가? True -> 사전생성이 들어간다.(형태소 쪼개고 인덱스로 사전구성)
tokenizer : 사전을 사용한다면 형태소(토큰) 분석기는 어떤 것이 담당할 것인가?
            기본값 : spacy, 한국어 -> tokenizer.morphs 를 사용
lower : 소문자로 처리해서 사용할 것인가?(대문자를 소문자로 처리하는 것이 영어권은 일반적)
batch_first : True -> [batch, h, w], False -> [h, w, batch] 
'''
TEXT = Field(
    sequential=True,            # 시퀀스 데이터이다.
    use_vocab=True,             # 사전을 사용하겠다.
    tokenize=tokenizer.morphs,  # 형태소 분석기로 이 함수를 사용하겠다.
    lower=True,                 # 소문자로 처리하겠다.
    batch_first=True            # 배치사이즈는 앞으로 위치
)

# 긍정/부정에 대한 필드
LABEL = Field(
    sequential=False,
    use_vocab=False,
    batch_first=True,
    preprocessing=lambda x: int(x),
    is_target=True
)

# 리뷰의 순서에 대한 필드(단, 사용하지는 않는다.)
# 데이터 로드 시 받아주기는 해야하므로 기본적인 것만 설정
ID = Field(
    sequential=False,
    use_vocab=False,
    is_target=False
)
```

## 데이터 셋 만들기

```
from torchtext.data import TabularDataset

# TabularDataset
# 데이터를 불러와서 정의한 필드를 기준으로 데이터를 처리해주는 객체
```

```
# 데이터를 주입하기 위해서 데이터를 준비한다.
dataset = TabularDataset(
    path='/content/ratings.txt', 
    format='tsv', 
    fields=[('id', ID), ('text', TEXT), ('label', LABEL)],
    skip_header=True
)

# 데이터를 읽어와서 정해진 규칙에 의해 전처리, 형태소 분리 등 처리하여 필드에 담는다.
```
