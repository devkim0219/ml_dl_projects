```
import keras
%tensorflow_version 1.x
```

# 간단한 데이터로 기본 루틴

```
import numpy as np
```

```
# 데이터 준비
x = np.array([0, 1, 2, 3, 4])

# 인공 신경망 : 딥러닝의 알고리즘이다.
# w(가중치) -> 2
# b(편향) -> 1
y = x * 2 + 1

x, y
```

```
# 인공 신경망 만들기 -> 층을 쌓는다.
# 1. 인공 신경망 모델을 만들기 시작했음을 파이썬 프로세스에게 알린다.
model = keras.models.Sequential()
```

```
# 2. 레이어 추가 -> 층을 쌓는다. -> add()
# 입력 노드는 1개, 가중치도 1개인 선형 처리 계층
model.add(keras.layers.Dense(1, input_shape=(1,)))
```

```
# 3. 학습단계 설정
# SGD : Stochastic Gradient Descent 경사 하강법(GD)
# mse : 손실함수를 표현할 때 사용하는 평균제곱오차(회귀) (Mean Squared Error)
model.compile('SGD', 'mse')
```

```
# 훈련
# 훈련 데이터, 답을 삽입
# epochs : 훈련을 인공 신경망이 한번 학습을 수행할 때 한 회, 한세대
# verbose : 로그, 학습 진행 상황을 표시, 0이면 미표시
model.fit(x[:2], y[:2], epochs=1000, verbose=0)
```

```
# 예측, 평가
model.predict(x[2:]).flatten()
```
