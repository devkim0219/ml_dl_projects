# 활성화 함수
- activation function
- 선형 방식(시스템)을 망에 적용하면, 망이 깊게 표현되지 않고 쉽게 예측이 된다. -> 딥하게 레이어를 구축하고 싶어도 이미 예측이 쉽게 되서 딥러닝을 구성할 수 없다.
- 딥러닝 환경에서 망이 깊어지면(레이어가 많아지면) -> 은닉층이 많아지게 되고 -> 매개변수가 줄어들고 연산 수가 줄어들어서 -> 정확도의 향상을 가져온다.
- 활성화 함수를 사용하는 이유
    - data를 비선형으로 바꾸기 위해서 사용된다.
    - 적절한 사용!!

|종류|이름|특징|
|--|--|:--|
|1|sigmoid|입력은 실수, 출력은 0 ~ 1사이|
|2|Tanh|입력은 실수, -1 ~ 1 사이로 변환|
|3|**ReLu**|**입력값이 음수면 0, 양수면 그대로 그 값**|
|4|Leaky ReLu|음수 입력 시 0.01과 같은 작은 수로 반환|
|5|PReLu|Leaky ReLu와 유사, 음수일 때 기울기 조정|
|6|ELU|음수 쪽 관련해서 y축에서 -1로 이동하여 0으로 근사하도록 미분처리한다.|
|7|Maxout|ReLu, Leaky ReLu의 일반화한 모델|

# sigmoid
- 실수를 취하고 0과 1사이의 숫자를 출력
- 음수는 클수록 0에 수렴, 양수는 클수록 1에 수렴
- 오랫동안 사용되어 왔으나, 큰 단점이 있어서 지금은 많이 사용이 안된다.
- 출력값이 1이나 0에 가까워지면 기울기가 0이되는 문제
- 이것은 가중치가 미반영 되어서 학습이 안되고 죽은 뉴런이라고 표현

# ReLu
- 최고의 인기, 가장 많이 사용
- ReLu, Leaky ReLu 등 렐루 계열을 가장 많이 사용
- 음수면 0, 양수면 자기 자신을 값으로 취한다.
- 장단점
    - 옵티마이저(경사 하강법 등...)가 더 빠르게 가중치를 찾도록 돕는 역할
    - SGD(확률적 경사 하강법)에서도 더 빠르게 수렴하도록 지원한다.
    - 연산 비용 저렴, 임계값을 가짐
    - 역전파 진행 시 기울기가 큰 값을 가진 경우, 응답이 없게 된다. -> 죽은 뉴런(학습율을 저하시키는 요소)
    - 이미지 인식대회에서 우승한 알고리즘 AlexNet에서 ReLu를 사용하여 Tanh보다 6배 성능향상을 가져왔다라고 논문에 기재
